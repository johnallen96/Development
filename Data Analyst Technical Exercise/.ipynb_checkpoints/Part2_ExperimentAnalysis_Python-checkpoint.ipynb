# Component 3 - Dependant Variable Analysis

### Step 1 - Variables

Now that we've concluded a measurement approach, we can assign our variables. 

Since we have our experiment groups defined as A and B, experiment_group will be our independent variable.

With our measurement approach concluded, we can assign num_exported_results as our dependent variable. 


```python
# Declare variables as dataframe col names
i_var = 'num_exported_results'
d_var = 'experiment_group'
baseline_rate = df[df[d_var] == 'A'][i_var].sum() / df[df[d_var] == 'A'][i_var].count()
practical_significance = 0.01 # we want to test for a 1% difference
confidence_level = 0.05 # for a 95% confidence interval
sensitivity = 0.8 
```

### Step 2 - Sampling the Population

A/B testing can become expensive when attempting to test an entire population of experiment data. The population needs to be an appropriate size that is large enough to produce a procise estimate but also small enough so that the analysis is practical and feasible to run.

A power analysis sample can be calculated to provide an estimated sample size of our data population that will serve us the requirements we need. 

```python
# Calculate required sample size
effect_size = sms.proportion_effectsize(baseline_rate, baseline_rate + practical_significance)
sample_size = sms.NormalIndPower().solve_power(effect_size = effect_size, power = sensitivity, 
                                               alpha = confidence_level, ratio=1)
print("Required sample size: ", round(sample_size), " per group")
```

```python
# Generate samples
control_sample = df[df['experiment_group'] == 'A'].sample(n=round(sample_size), random_state=22)
treatment_sample = df[df['experiment_group'] == 'B'].sample(n=round(sample_size), random_state=22)

# Concatenate back into a single dataframe
df = pd.concat([control_sample, treatment_sample], axis=0)
df.reset_index(drop=True, inplace=True)
```

```python
export_rates = df.groupby(d_var)[i_var]

export_rates = export_rates.agg([np.sum, np.mean, np.std])
export_rates.columns = ['sum_exports', 'export_rate', 'std_deviation']

export_rates.style.format('{:.3f}')
```

To see if there's a visual difference between these export rates, we can plot the data as a barplot.

```python
plt.figure(figsize=(8,6))

sns.barplot(x=df[d_var], y=df[i_var], ci=False)

plt.title('Export Rate by Experiment Group')
plt.xlabel('Experiment Group')
plt.ylabel('Export Rate')
plt.show()
```

As we can see above, experiment group A have a slightly higher export rate. Now we can test if this difference is statistically significant and rejects a null hypothesis. 

A null hypothesis is where a control group rate equals a treatment group rate to support statistical insignificance between both groups. If the difference between these rates is greater than the practical significance set then the null hypothesis can be suggested and the alternate hypothesis is supported.

These are: 

 - H0: p = p0
 - Ha; p != p0
 
Where p and p0 represent the measure rate of the control and treatment groups respectfully. In this test, Group A is our control group and Group B is our treatment group. 

### Step 3 - A/B Test

```python
# Define power factors
power=0.8
alpha=0.05
ratio=1

# Split dataframe
control_results = df[df[d_var] == 'A'][i_var]
treatment_results = df[df[d_var] == 'B'][i_var]
```

```python
# Calculate Pooled Probability

control_results = df[df[d_var] == 'A'][i_var]
exports_control = control_results.sum()
total_records_control = control_results.count()

treatment_results = df[df[d_var] == 'B'][i_var]
exports_treatment = treatment_results.sum()
total_records_treatment = treatment_results.count()

baseline_rate = exports_control/total_records_control
practical_significance = 0.01 # - for a 1% change significance in either direction of the control
confidence_level = 0.05 # - for a 95% confidence interval
sensitivity = 0.8

prob_pooled = (exports_control + exports_treatment) / (total_records_control + total_records_treatment)
```

```python
# Calculate pooled standard error and margin of error
se_pooled = math.sqrt(prob_pooled * (1 - prob_pooled) * (1 / total_records_control + 1 / total_records_treatment))
z_score = sts.norm.ppf(1 - confidence_level / 2)
margin_of_error = se_pooled * z_score

# Calculate dhat, the estimated difference between probability of exports in the experiment and control groups
d_hat = (exports_treatment / total_records_treatment) - (exports_control / total_records_control)

# Test if we can reject the null hypothesis
lower_bound = d_hat - margin_of_error
upper_bound = d_hat + margin_of_error
```

```python
# Convert negative number into positive number for change significance test
#  as we're testing a change in either direction of the control
if practical_significance < abs(lower_bound):
    print('Conclusion: Reject null hypothesis')
else: 
    print('Conclusion: Do not reject the null hypothesis')
    
print('The lower bound of the confidence interval is ', round(lower_bound * 100, 2), "%")
print('The upper bound of the confidence interval is ', round(upper_bound * 100, 2), "%")
```

# Component 4 - Final Recommendation

As we can see that our treatment group has a lower export rate to our control group, from the sampled ab_test that we've performed, the lower bound of the confidence interval is a change rate difference of 0.02%. 

Because this is so close, it could be argued that our sampling could have returned a slighly wider distribution than expected, and could be causing an broader and therefore less accurate calculation. As always in statistics, the more data you have, the narrower the distribution and the more precise the estimate. Under a full analysis, we would reduce random chance like this by making small adjustments the sample size to test the statistical inference against a narrower distribution. We could also bootrap the dataset to continously simulate random sampling from an evolving population. This can be effective when we are unable to gather sufficient data points for a precise estimate. 

Moving forward, since we've concluded that our conficence interval is between -0.98% and 0.41%, we could only reject the null hypothesis if the confidence interval lower bound was above 1%; where 1% is the practical significance of change that we declared. Since we marginally fell short by 0.02%, we are unable to reject the null hypotheis and can conclude that the effect of experiment group B is statistically equivalent to the effect of experiment group A. 

In the context of business, this supports that automated elevations are statistically just as effective as manual elevations from the objective of image exports. When evaluating the business need of these elevation methods, assuming we're scoped to the single measure that we selected and that all other factors are excluded, the process of maintaining manual elevations does not appear to deliver a significant value add that justifies it's inclusion for search optimisation.
